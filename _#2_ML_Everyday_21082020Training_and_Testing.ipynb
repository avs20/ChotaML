### Training and Testing Dataset

Suppose we give all our dataset to train (finding the algorithm)

What happens when we try to check how it's working. Do we have new data ? No 
So we will check with the data we passed during training and here what the algo can do. It can just learn the values. If B is 0 return 0 or if B is 1 return 1.

So we can never know if the algo has just learned the values or did some work and learned the algorithm. Just like in class. We do one type of question and then test on another question of same type. If we just learn the value, the answer will be wrong. But if we learn the method , the answer will be correct. 

So we divide the data in 2 parts. 
1. Training data set 
2. Testing data set. 

If the algorithm has genuinely learned, it will work good on the testing dataset also. 




# Things to keep in mind 

1. Test data set should not be part of testing data set 
2. It should be representative of the training data set. 
3. Size can be based on the data, for DL and huge dataset, 1 % might be enough. 
4. For smaller datset, 30-70 or 80-20 is used. 
5. The accuracy of the model is on the test dataset and not training 
6. Accucay = High on training and low in testing tells overfitting. 
7. Low in training and Low in testing, tells, not good enough and something should be changed. (What? it depends)
8. For large dataset, there is another division, known as validation set.
9. How to divide data in test and train, depends on problem
